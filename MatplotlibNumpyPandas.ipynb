**Matplotlib**

```python
from matplotlib import pyplot as plt
x=[2000,2005,2010,2015]
y=[2,4,6,8]
plt.plot(x,y,color='red')
plt.xlabel("Year")
plt.ylabel("Population")
plt.title("City")
plt.yticks([0,2,4,6,8],['0M','2M','4M','6M','8M'])
plt.grid()
plt.show()
```

```python
categories = ['A', 'B', 'C', 'D']
values = [10, 15, 8, 12]

plt.bar(categories, values)
plt.title("Bar Plot")
plt.xlabel("Category")
plt.ylabel("Value")
plt.show()
```

```python
import numpy as np

x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 2, 1, 3, 5, 4])

plt.plot(x, y)
plt.title("Line Plot")
plt.xlabel("X-axis Label")
plt.ylabel("Y-axis Label")
plt.show()
```

**Numpy**

```python
my_array=np.array([[1,2,3,4,5],[11,11,11,11,11],[1,1,1,1,1]])
print(my_array)
np.column_stack(my_array)       # column_stack is a function in the NumPy library used to stack 1-D arrays as columns into a single 2-D array
```

```python
my_array2=np.array([[1,2,3],[2,2,2],[1,1,1]])
np.mean(my_array2[2,:])
```

```python
my_array3=np.column_stack([[1,2,3,4],[1,1,1,1]])
print(my_array3)
print(np.mean(my_array3[:,0]))
print(np.mean(my_array3[:,1]))
```

```python
array=np.array([9,8,7,6,5,4])
for i in enumerate(array):
  print(i)
```

```python
array=np.array([[1,2,3],[1,2,3]])
for i in (array):
  print(i)
```

```python
array=np.array([[1,2,3],[1,2,3]])
for i in np.nditer(array):
  print(i)
```

```python
import pandas as pd
dict={'Name':['Ali','Ahmed','Hadi', 'Waqas'], 'Country':['PAK','ITA','KSA','PAK'],'Age':[27,26,26,26]}
df=pd.DataFrame(dict)
print(df)
```

```python
print(df.iloc[[0],[0]])   # .iloc (Integer-location based indexing)
```

```python
print(df.iloc[[0],[1]])
```

```python
print(df.iloc[[1],[0]])
```

```python
print(df.iloc[0,0])
```

```python
print(df.iloc[0,1])
```

```python
print(df.iloc[0,2])
```
