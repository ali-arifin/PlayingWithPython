**Understanding Recursion**

```python
def factorial(n):
  print("Factorial called with " + str(n))
  if n < 2:
    print("Returning 1")
    return 1
  result = n * factorial(n-1)
  print("Returning " + str(result) + " for factorial of " + str(n))
  return result

factorial(4)
```

**Let’s follow what happens when we call factorial(4):**

factorial(4)

→ prints: Factorial called with 4
→ calls factorial(3)

factorial(3)

→ prints: Factorial called with 3
→ calls factorial(2)

factorial(2)

→ prints: Factorial called with 2
→ calls factorial(1)

factorial(1)

→ prints: Factorial called with 1

→ hits base case → prints: Returning 1

→ returns 1

**Now the recursive calls return step-by-step:**

factorial(2) = 2 * 1 = 2

→ prints: Returning 2 for factorial of 2

factorial(3) = 3 * 2 = 6

→ prints: Returning 6 for factorial of 3

factorial(4) = 4 * 6 = 24

→ prints: Returning 24 for factorial of 4
